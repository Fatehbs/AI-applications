```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#Traits
man_trait = [['Rich', 'Poor'], ['Strong', 'Weak'], ['Intelligent', 'Dumb'], ['Loyal', 'Cheat']]
woman_trait = [['Beautiful', 'Ugly'], ['Caring', 'Toxic'], ['Creative', 'Dull'], ['Loyal', 'Cheat']]
sigma = 2
```

```python
man_payoff = pd.DataFrame(np.zeros((16, 16)), index = pd.MultiIndex.from_product(man_trait), columns = pd.MultiIndex.from_product(woman_trait))

man_payoff['Beautiful'] += 1
man_payoff.loc[:, pd.IndexSlice[:, 'Caring']] += 1
man_payoff.loc[:, pd.IndexSlice[:, :, 'Creative']] += 1
man_payoff.loc[:, pd.IndexSlice[:, :, :, 'Loyal']] += 1
```

```python
woman_payoff = pd.DataFrame(np.zeros((16, 16)), index = pd.MultiIndex.from_product(woman_trait), columns = pd.MultiIndex.from_product(man_trait))

woman_payoff['Rich'] += 1
woman_payoff.loc[:, pd.IndexSlice[:, 'Strong']] += 1
woman_payoff.loc[:, pd.IndexSlice[:, :, 'Intelligent']] += 1
woman_payoff.loc[:, pd.IndexSlice[:, :, :, 'Loyal']] += 1
```

```python
man_satisfaction = man_payoff - woman_payoff.T
woman_satisfaction = woman_payoff - man_payoff.T
```

```python
woman_satisfaction
```

```python
man_reward = man_payoff + man_satisfaction
woman_reward = woman_payoff + woman_satisfaction
```

```python
woman_reward
```

```python
#Payoff Matrix of Player A
A = man_reward.values

#Payoff Matrix of Player B
B = woman_reward.values
```

```python
pA = np.ones((16,)) / 16
pB = np.ones((16,)) / 16

pA_next = pA.copy()
pB_next = pB.copy()
```

```python
def step_A():
    global pA_next
    fA = np.matmul(A, pB)
    fA_av = np.dot(pA, fA)

    mu_A = fA - fA_av
    pA_next = pA * np.exp(mu_A)

    pA_next = pA_next / np.sum(pA_next)
```

```python
def step_B():
    global pB_next
    fB = np.matmul(B, pA)
    fB_av = np.dot(pB, fB)

    mu_B = fB - fB_av
    pB_next = pB * np.exp(mu_B)

    pB_next = pB_next / np.sum(pB_next)
```

```python
def step():
    global pA
    global pB

    step_A()
    step_B()

    pA = pA_next
    pB = pB_next
```

```python
for _ in range(10):
    step()
```

```python
prob_df = pd.DataFrame(np.reshape(pA, (16, 1)) * np.reshape(pB, (1, 16)),
    index = man_reward.index,
    columns = man_reward.columns
)

prob_df.index.name = 'Man'
prob_df.columns.name = 'Woman'
```

```python
g = sns.heatmap(prob_df)

g.set_xlabel("Man")
g.set_ylabel("Woman")
```
