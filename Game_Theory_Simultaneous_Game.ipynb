```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
reward_arr = np.array([
    [[2, 1], [0, 0]],
    [[0, 0], [1, 2]]
])

players = ['husband', 'wife']
stategies = ['cricket', 'movie']
```

```python
husband_rewards_df = pd.DataFrame(
    data = reward_arr[..., 0],
    columns = stategies,
    index = stategies
)

wife_rewards_df = pd.DataFrame(
    data = reward_arr[..., 1],
    columns = stategies,
    index = stategies
).T
```

```python
husband_rewards_df
```

```python
wife_rewards_df
```

```python
husband_df = pd.DataFrame({
    'prior' : pd.Series([0.5, 0.5], index = stategies),
    'pred' : pd.Series([0, 0], index = stategies),
    'post' : pd.Series([0, 0], index = stategies)
})

wife_df = pd.DataFrame({
    'prior' : pd.Series([0.5, 0.5], index = stategies),
    'pred' : pd.Series([0, 0], index = stategies),
    'post' : pd.Series([0, 0], index = stategies)
})
```

```python
husband_tw = husband_rewards_df.apply(np.exp)
husband_tp = husband_tw / husband_tw.sum()

wife_tw = wife_rewards_df.apply(np.exp)
wife_tp = wife_tw / wife_tw.sum()
```

```python
for _ in range(30):
    husband_df['pred'] = np.matmul(husband_tp, wife_df['prior'])
    husband_df.eval('post = prior * pred', inplace = True)
    husband_df['post'] /= husband_df['post'].sum()

    wife_df['pred'] = np.matmul(wife_tp, husband_df['prior'])
    wife_df.eval('post = prior * pred', inplace = True)
    wife_df['post'] /= wife_df['post'].sum()

    husband_df.eval('prior = post', inplace = True)
    wife_df.eval('prior = post', inplace = True)
```

```python
results_df = pd.DataFrame({
    'husband' : husband_df['post'],
    'wife' : wife_df['post']
})
```
